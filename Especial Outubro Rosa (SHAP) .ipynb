```python
import pandas as pd
import numpy
%matplotlib inline
```

```python
data = pd.read_csv("data/data.csv")
```

```python
data.head()
```

```python
# Filter by mean columns
features = data.filter(regex=r"_mean", axis=1)
```

```python
features.head()
```

```python
y = data['diagnosis'].map({"M": 0, "B": 1})
```

```python
y.value_counts()
```

```python
features.shape, y.shape
```

### Division of training and testing databases

```python
from sklearn.model_selection import train_test_split

Xtrain, Xval, ytrain, yval = train_test_split(features, y, train_size=0.5)
```

```python
Xtrain.shape, Xval.shape, ytrain.shape, yval.shape
```

### Create Random Forest Classifier Model 

```python
from sklearn.ensemble import RandomForestClassifier

mdl = RandomForestClassifier(n_estimators=1000, 
                             n_jobs=-1, 
                             random_state=0)

# mdl.fit(Xtrain, ytrain)
```

```python
from timeit import default_timer as timer
from datetime import timedelta

# Start Measure time elapsed
start = timer()
# Code here ...

mdl.fit(Xtrain, ytrain)

# End Measure time elapsed
end = timer()
print(timedelta(seconds = end - start))
```

### Predictions and evaluations

```python
from sklearn.metrics import classification_report
p = mdl.predict(Xval)
print(classification_report(yval, p))
```

## SHAP

```python
import shap

# load JS visualization code to notebook
shap.initjs()
```

```python
p[:5]
```

```python
explainer = shap.TreeExplainer(mdl)
shap_values = explainer.shap_values(Xtrain)
```
